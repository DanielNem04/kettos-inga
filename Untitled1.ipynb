```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
import matplotlib as mpl

# 🔧 FFMPEG elérési út beállítása – ezt futtasd a terminálban is: which ffmpeg
mpl.rcParams['animation.ffmpeg_path'] = '/opt/homebrew/bin/ffmpeg'

# 🔄 CSV beolvasás
df = pd.read_csv("build/double_pendulum_output.csv", sep=r'\s+|,|\t+', engine='python')
df.columns = df.columns.str.strip()

# ⚠️ Oszlopellenőrzés
required = ['time', 'x1', 'y1', 'x2', 'y2']
for col in required:
    if col not in df.columns:
        raise ValueError(f"Hiányzó oszlop: {col}")

# 🔍 Tengelyhatárok
x_min = min(df['x1'].min(), df['x2'].min()) - 0.5
x_max = max(df['x1'].max(), df['x2'].max()) + 0.5
y_min = min(df['y1'].min(), df['y2'].min()) - 0.5
y_max = max(df['y1'].max(), df['y2'].max()) + 0.5

# 🎨 Ábra beállítás
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
ax.set_xlabel(r'$x$ (m)')
ax.set_ylabel(r'$y$ (m)')
ax.set_title('Kettős inga animáció')
ax.grid(True)
ax.set_aspect('equal')

# 🔧 Rajzelemek
line1, = ax.plot([], [], 'o-', lw=2, color='blue')
line2, = ax.plot([], [], 'o-', lw=2, color='red')
trace, = ax.plot([], [], '-', lw=1, color='gray', alpha=0.6)
time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)

# 🧠 Nyomvonal-tárolók
x_trail, y_trail = [], []

def init():
    line1.set_data([], [])
    line2.set_data([], [])
    trace.set_data([], [])
    time_text.set_text('')
    return line1, line2, trace, time_text

def update(i):
    x0, y0 = 0, 0
    x1, y1 = df.loc[i, 'x1'], df.loc[i, 'y1']
    x2, y2 = df.loc[i, 'x2'], df.loc[i, 'y2']

    line1.set_data([x0, x1], [y0, y1])
    line2.set_data([x1, x2], [y1, y2])

    x_trail.append(x2)
    y_trail.append(y2)
    trace.set_data(x_trail, y_trail)

    time_text.set_text(f"t = {df.loc[i, 'time']:.2f} s")
    return line1, line2, trace, time_text

# 🎥 Animáció generálása
ani = FuncAnimation(fig, update, frames=len(df), init_func=init, blit=True)

# 💾 Mentés MP4-be
writer = FFMpegWriter(fps=30, metadata=dict(artist='double_pendulum'))
ani.save("double_pendulum_animation.mp4", writer=writer)

print("✅ Animáció sikeresen mentve: double_pendulum_animation.mp4")
```
